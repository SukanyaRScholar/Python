Question 2: Write a Python program to sort a list alphabetically in a dictionary.

```
dic={'a':['l','e','a','r','n'], 'b':['p','y','t','h','o','n'], 'c':['S','h','i','v','a','m'], 'd':['r','e','s','t']}
# program to sort the list of values in a dictionary
sorteddict={key: sorted(value) for key, value in dic.items()}
print(sorteddict)

# program to sort the items in a dictionary based on keys and not values
dic={'a':['l','e','a','r','n'], 'c':['p','y','t','h','o','n'], 'd':['S','h','i','v','a','m'], 'b':['r','e','s','t']}
sortedkeys={k:v for k,v in sorted(dic.items())}
print(sortedkeys)

# program to sort the list based on values only and not the key
diction={'a':'Sukanya', 'd':'Learning', 'b': 'Is', 'c':'Python'}
sortedvalues={k:v for k,v in sorted(diction.items(),key= lambda v: v[1])}
print(sortedvalues)
```

Question 4:
(i) Write a Python Program to implement your own myreduce () function which
works exactly like Python's built-in function reduce ().
(ii) Write a Python program to implement your own myfilter () function which works
exactly like Python's built-in function filter ().

```
def myreduce(lists, operation):
  reduced=0
  
  for i in lists:
    if operation=='add':
      reduced+=i
    elif operation=='subtract':
      reduced-=i

  return reduced

def myfilter(lists):
  filtered=[]
  for i in lists:
    if i%2==0:
      filtered.append(i)

  return filtered

print(myreduce([1,2,3,4,5], 'add'))
print(myfilter([1,2,3,4,5]))
```
